```python
from skimage.feature import local_binary_pattern
import os
import cv2
import numpy as np

# Image preprocessing functions
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    
    # Check if the image loaded successfully
    if img is None:
        return None
    
    # Non-Local Means Denoising
    img = cv2.fastNlMeansDenoisingColored(img, None, 10, 10, 7, 21)
    
    # Convert the image to grayscale for CLAHE
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply CLAHE
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    cl1 = clahe.apply(gray)
    
    # Resize the image
    img_resized = cv2.resize(cl1, (img_size, img_size))  
    return img_resized


def calculate_lbp_features(img):
    lbp_radius = 3
    lbp_points = 8 * lbp_radius
    lbp = local_binary_pattern(img, lbp_points, lbp_radius, method='uniform')
    return lbp.flatten()

def extract_sift_features(img):
    sift = cv2.SIFT_create()
    _, descriptors = sift.detectAndCompute(img, None)  # use img directly as it's already grayscale
    if descriptors is None:
        descriptors = np.zeros((1, 128))
    
    # Flatten and ensure the descriptor is of length 1000
    descriptors_flat = descriptors.flatten()
    
    # Check the length of descriptors_flat before padding or truncating
    if len(descriptors_flat) >= 1000:
        padded_descriptors = descriptors_flat[:1000]
    else:
        padded_descriptors = np.pad(descriptors_flat, (0, 1000 - len(descriptors_flat)), mode='constant')
    
    return padded_descriptors

# New Path: "/Users/ummefahmidaakter/Desktop/Completed Assessments/Dissertation/dissertation_project/up4"
# If any error occurs while running this code, it mostly could be bacause of incorrect root path

img_size = 250
hist_bins = 16
train_folder = "/Users/ummefahmidaakter/Downloads/up4/train"
test_folder = "/Users/ummefahmidaakter/Downloads/up4/test"
val_folder = "/Users/ummefahmidaakter/Downloads/up4/val"
classes = ['polyp', 'uterus']

lbp_features_train = []
sift_features_train = []
labels_train = []

lbp_features_test = []
sift_features_test = []
labels_test = []

lbp_features_val = []
sift_features_val = []
labels_val = []

# Loop through each class in the dataset
for c in classes:
    # Train data
    folder_train = os.path.join(train_folder, c)
    for file in os.listdir(folder_train):
        file_path = os.path.join(folder_train, file)
        label = file_path.split("/")[-2]
        img = preprocess_image(file_path)
        
        # Check if the image preprocessing was successful
        if img is not None:
            lbp = calculate_lbp_features(img)
            descriptors = extract_sift_features(img)
            lbp_features_train.append(lbp)
            sift_features_train.append(descriptors)
            labels_train.append(label)

    # Test data
    folder_test = os.path.join(test_folder, c)
    for file in os.listdir(folder_test):
        file_path = os.path.join(folder_test, file)
        label = file_path.split("/")[-2]
        img = preprocess_image(file_path)
        
        # Check if the image preprocessing was successful
        if img is not None:
            lbp = calculate_lbp_features(img)
            descriptors = extract_sift_features(img)
            lbp_features_test.append(lbp)
            sift_features_test.append(descriptors)
            labels_test.append(label)

    # Validation data
    folder_val = os.path.join(val_folder, c)
    for file in os.listdir(folder_val):
        file_path = os.path.join(folder_val, file)
        label = file_path.split("/")[-2]
        img = preprocess_image(file_path)
        
        # Check if the image preprocessing was successful
        if img is not None:
            lbp = calculate_lbp_features(img)
            descriptors = extract_sift_features(img)
            lbp_features_val.append(lbp)
            sift_features_val.append(descriptors)
            labels_val.append(label)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix

# Combine LBP and SIFT features
train_features = np.concatenate((lbp_features_train, sift_features_train), axis=1)
test_features = np.concatenate((lbp_features_test, sift_features_test), axis=1)
val_features = np.concatenate((lbp_features_val, sift_features_val), axis=1)

# Initialize MLP classifier
mlp = MLPClassifier(max_iter=1000, early_stopping=True, n_iter_no_change=10)

# Define hyperparameter grid for tuning
param_grid = {
    'hidden_layer_sizes': [(100,), (200,), (300,)],
    'alpha': [0.0001, 0.001, 0.01],
    'activation': ['logistic', 'relu']
}

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(mlp, param_grid, cv=3, verbose=2)
grid_search.fit(val_features, labels_val)  # Using validation dataset for tuning

# Store the best hyperparameters
best_params = grid_search.best_params_
print("Best hyperparameters:", best_params)
```

```python
# Use the best MLP model from GridSearchCV
best_mlp = grid_search.best_estimator_

# Evaluate the model on the testing set
predicted_labels = best_mlp.predict(test_features)

accuracy = accuracy_score(labels_test, predicted_labels)
precision = precision_score(labels_test, predicted_labels, average='weighted', zero_division=1)
recall = recall_score(labels_test, predicted_labels, average='weighted')
f1 = f1_score(labels_test, predicted_labels, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)

# Calculate and print confusion matrix
conf_matrix = confusion_matrix(labels_test, predicted_labels)
print("\nConfusion Matrix:\n", conf_matrix)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, auc
import matplotlib.pyplot as plt

# Assuming the classes are 'polyp' and 'uterus', we need to convert these string labels to binary format (0s and 1s)
binary_labels_test = [0 if label == 'polyp' else 1 for label in labels_test]

# Get the probability estimates of the positive class (in this case, 'polyp')
probabilities = best_mlp.predict_proba(test_features)[:, 1]

# Compute the ROC curve
fpr, tpr, thresholds = roc_curve(binary_labels_test, probabilities)

# Compute the AUC
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
